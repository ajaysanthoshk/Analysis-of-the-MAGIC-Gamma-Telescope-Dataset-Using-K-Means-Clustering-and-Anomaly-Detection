```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Load the dataset
file_path = 'magic04.data'
columns = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1', 'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist', 'Class']
data = pd.read_csv(file_path, header=None, names=columns)

# Data Preprocessing
# Map 'g' to 0 and 'h' to 1 for the 'Class' column
data['Class'] = data['Class'].map({'g': 0, 'h': 1})

# Separate features and target
X = data.drop('Class', axis=1)
y = data['Class']

# Normalize the features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Evaluate K-Means
silhouette_avg = silhouette_score(X_scaled, clusters)
print(f"Silhouette Score for K-Means: {silhouette_avg}")

# Add cluster labels to the data
data['Cluster'] = clusters

# Clustering-Based Anomaly Detection
# Calculate distance of each point to its cluster center
data['DistanceToCentroid'] = np.linalg.norm(X_scaled - kmeans.cluster_centers_[clusters], axis=1)

# Define anomaly threshold (e.g., top 5% distances)
thresh = np.percentile(data['DistanceToCentroid'], 95)
data['Anomaly'] = data['DistanceToCentroid'] > thresh

# Summary of anomalies
anomalies = data[data['Anomaly']]
print(f"Number of anomalies detected: {len(anomalies)}")

# Visualizations
# Plot Cluster Distribution
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Plot Distance to Centroid Distribution
plt.figure(figsize=(8, 6))
plt.hist(data['DistanceToCentroid'], bins=30, color='blue', alpha=0.7)
plt.axvline(thresh, color='red', linestyle='dashed', linewidth=1, label='Anomaly Threshold')
plt.title('Distance to Centroid')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Save anomalies to a CSV file
anomalies.to_csv('anomalies.csv', index=False)
print("Anomalies saved to anomalies.csv")
```
